### Matplotlic이란?
- 파이썬에서 데이터를 차트나 플롯으로 그려주는 시각화를 위한 라이브러리
- 차트의 종류는 산점도, 바, 파이, 히스토그램, 꺽은선 그래프 등 다양하게 존재함

```python
# matplolib 불러오기
import matplotlib.pyplot as plt
```

#### 기본적인 그래프 그려보기

```python
# x, y 축 지정해서 선 그래프를 그려보자!
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y) # plot : 선 그래프를 그려주는 명령
plt.show()    # 그래프를 출력
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls='--', lw= 5) 
plt.show() 
```

```python
# x, y 축 지정해서 선 그래프를 그려보자!
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls='--', lw=3, marker='o') # plot : 선 그래프를 그려주는 명령
plt.show()    # 그래프를 출력
```

```python
# x, y 축 지정해서 선 그래프를 그려보자!
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls='--', lw=3, marker='o', c='y') # plot : 선 그래프를 그려주는 명령
plt.show()    # 그래프를 출력
```

```python
# x, y 축 지정해서 선 그래프를 그려보자!
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls='--', lw=3, marker='o', c='y', ms=20) # plot : 선 그래프를 그려주는 명령
plt.show()    # 그래프를 출력
```

### 차트 옵션

#### 차트 내부의 그림의 범위를 지정

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', marker='o', mfc='r', lw=5, c='b', mew=5, mec= "g" , ms=15)
plt.xlim(-2, 8)
plt.xlim(-5, 12)
plt.show() 
```

#### 틱 설정
- 그래프에서 x, y축 상의 위치 표시 지점을 틱(tick)이라고 함
- 틱은 일반적으로 Matplotlib 에서 자동으로 정해주지만 수동으로 설정 할 수 있음

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', marker='o', mfc='r', lw=5, c='b', mew=5, mec= "g" , ms=15)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show() 
```

#### 그리드
- 그래프에서 틱의 위치를 잘 보여주기 위하여 그림 중간에 선이 자동으로 그려지는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', marker='o', mfc='r', lw=5, c='b', mew=5, mec= "g" , ms=15)
plt.grid()
plt.show() 
```

#### 여러개의 그래프를 한번에 그리기

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
x1 = np.arange(10)
y1 = np.arange(10)
# 1번째 플롯
plt.plot(x,y, ls='--',  c='g', marker='o',ms=5)
# 2번째 플롯
plt.plot(x1, y1)

plt.show() 
```

#### 요일별 사고 건수 카운트

```python
data
```

```python
# 요일 컬럼의 유니크값과 각각의 개수를 출력
temp = data['요일'].value_counts()    # 기본적으로 내림차순
temp
```

```python
temp = data['요일'].value_counts()  
y = temp[['월','화','수','목','금','토','일']]   # 유니크값으로 설정
y
```

#### 전국 사망 교통사고 시각화

```python
x = range(7)
plt.bar(x, y)
plt.xticks(x, ['M', 'T', 'W', 'TH', 'F', 'ST', 'S'])
plt.show()
```

```python
# Y축 범위 : 500~620
# X, Y축 제목 설정
# TITLE  : 전체 그래프 제목 설정
x = range(7)
plt.bar(x, y)
plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])
plt.ylim(500, 620)
plt.xlabel("요일")
plt.ylabel("사고 건수")
plt.title("요일별 사망교통사고 건수")
plt.show()
```
