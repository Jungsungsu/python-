- 함수란?
- 하나의 특별한 목적의 작업을 수행하기 위해 독립적으로 설계된 코드의 집합

- 함수를 사용하는 이유
- 반복적인 프로그래밍을 피할 수 있다.
- 모듈화로 인해 전체적인 코드의 가독성이 좋아진다.
- 프로그램에 문제가 발생하거나 기능의 변경이 필요할 때에도 손쉽게 유지보수가 가능하다.

- def 함수명(매개변수):
    - 실행문장
    - return 반환변수

```python
def number_sum(num1, num2):
    result = num1 + num2
    return result
```

```python
number_sum(3,10)
```

```python
number = number_sum(3,10)
number
```

```python
def number_minus(num1, num2):    
    if num1 >= num2:
        result = num1- num2
    else:
        result = num2-num1
        
    return result
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))

result = number_minus(num1, num2)
result
```

```python
def s_replace(word):
    return word.replace("ㅋ", "")
```

```python
s = input("문자열 입력 >> ")
result = s_replace(s)
result
```

```python
def cal(num1, num2, op): 
    """설명을 작성해주세요"""
    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    return result
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >> ")
result = cal(num1, num2, op)
print(f"결과 : {result}")
```

```python
#리뷰의 전체 길이를 구해서 개수로 나눠주면 됨
total = 0 #리뷰의 길이가 더해져서 담기게 될 변수
for i in my:
    total += len(i[0])
print(f" 리뷰의 평균 길이는 '{total/len(my)}' 입니다.")
```

```python
# 첫번째 방법
neg_cnt = 0   #부정 리뷰의 개수
pos_cnt = 0   #긍정 리뷰의 개수

for i in my :
    #리뷰가 부정인 경우
    if i[1] == 0 :
        neg_cnt += 1
    #리뷰가 긍정인 경우
    else :
        pos_cnt += 1
print(f"긍정 '{pos_cnt}'개")
print(f"부정 '{neg_cnt}'개")
```

```python
# 두번째 방법
re_list = []   # 0 또는 1 값을 다 넣어줄 공간

for i in my :
    re_list.append(i[1])

neg = re_list.count(0)    # 0이라는 숫자를 카운트
pos = re_list.count(1)    # 1이라는 숫자를 카운트

print(f"긍정 '{pos}'개")
print(f"부정 '{neg}'개")
```

```python
cnt = 0
data = input("검색할 데이터 입력 >> ")
for i in my:
    if data in i[0] :  # 리뷰 내에서 입력한 문자열이 있는지
        cnt += 1
print(f"'{data}'가 포함된 리뷰는 {cnt}개 입니다.")
```
